```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots to be created


# Read in the data

df = pd.read_csv("/Users/kaciemorinaga/Downloads/movies.csv")
```

```python
# Take a look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop rows with missing data

df = df.dropna()
```

```python
# Check if rows were dropped

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types

df.dtypes
```

```python
# Convert data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Check for changes

df.head()
```

```python
# Correcting year to match year in released column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Sorting by gross

df_sorted = df.sort_values(by=["gross"], inplace=False, ascending=False)
```

```python
#Look at all of the data

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatterplot with budget vs. gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs. Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs. gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
# Looking at correlations

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
# Show heatmap for quick visualization

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df.head()
```

```python
# Numerize features to look for other correlations

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized = df_numerized.sort_values(by=["gross"], inplace=False, ascending=False)
        
df_numerized.head()
```

```python
# Compare numerized to nominal

df_sorted.head()
```

```python
# New heatmap with numerized features

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at the table

df_numerized.corr()
```

```python
# Looking feature by feature

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sorting by highest to lowest correlations

sorted_pairs = corr_pairs.sort_values(ascending = False)

sorted_pairs
```

```python
# Showing only correlations greater than 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
